Importing libz

```
import keras
```

```
import numpy as np
```

Defining model

```
my_dummy_model = keras.Sequential([keras.layers.Dense(units=1, input_shape = [1])])
```

Compiling model with number of epochs

```
my_dummy_model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
X = np.array([-1.,0.,1.,2.,3.,4.], dtype=float)
y = np.array([-3.,-1.,1.,3.,5.,7.], dtype=float)
```

Training! :)

```
my_dummy_model.fit(X, y, epochs=500)
```

Predicting an unkown value

```
my_dummy_model.predict([5.])
```
